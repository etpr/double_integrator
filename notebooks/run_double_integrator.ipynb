```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append("../")
from double_integrator import DoubleIntegrator
plt.rcParams['figure.figsize'] = [10, 10]
%load_ext autoreload
%autoreload 2
```

```python
x_0 = np.array([0.5, 1.0])
x_f = np.array([-0.5, 1.0])

di = DoubleIntegrator(u_sys_lim=1.0, dt=0.001)
res, traj = di.time_opt_control(x_0=x_0, x_f=x_f)


plt.figure(1)
traj.plot_traj()

plt.figure(2)
traj.plot_phase()
plt.show()
```

```python
di = DoubleIntegrator(u_sys_lim=1, dt=0.01)
n = 5
x, y = np.meshgrid(np.linspace(-1, 1, n), np.linspace(-1, 1, n))
z = np.transpose(np.vstack((x.flatten(), y.flatten())))

plt.figure(1)
for x in np.row_stack(z):
    x_0 = np.array([x[0], x[1]])
    x_f = np.array([0.0, 0.0])
    
    res, traj = di.time_opt_control(x_0=x_0, x_f=x_f)
    traj.plot_phase(resize=False)
    
plt.xlim([-2, 2])
plt.ylim([-2, 2])
ax = plt.gca()
ax.set_aspect('equal')
ax.grid(True, which='both')
plt.xlabel('x')
plt.ylabel('$\dot{x}$')
    
plt.show()
```
