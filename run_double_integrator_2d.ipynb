```python
import matplotlib.pyplot as plt
import numpy as np
from random import random, seed
from double_integrator_2d import DoubleIntegrator2D
```

```python
x_10 = 0.0
x_20 = 0.5
x_1f = 1.0
x_2f = 0.0
y_10 = 0.0
y_20 = 0.0
y_1f = 1.0
y_2f = 0.5

di = DoubleIntegrator2D(u_sys_lim=1.0, dt=0.01)
res, traj = di.steer(x_10=x_10, x_20=x_20, x_1f=x_1f, x_2f=x_2f, y_10=y_10, y_20=y_20, y_1f=y_1f, y_2f=y_2f)
if res:
    [t_f_x, t_s_x, y_1s, y_2s, X, U_x, S_x, T_x, t_f_y, t_s_y, y_1s, y_2s, Y, U_y, S_y, T_y] = traj
    
    plt.figure(1)
    ax = plt.subplot(411)
    ax.grid(True, which='both')
    plt.ylabel('x_1')
    plt.plot(T_x, X[:, 0])
    plt.plot(T_y, Y[:, 0])
    ax = plt.subplot(412)
    ax.grid(True, which='both')
    plt.ylabel('x_2')
    plt.plot(T_x, X[:, 1])
    plt.plot(T_y, Y[:, 1])
    ax = plt.subplot(413)
    ax.grid(True, which='both')
    plt.plot(T_x, U_x)
    plt.plot(T_y, U_y)
    plt.ylabel('u')
    ax = plt.subplot(414)
    ax.grid(True, which='both')
    plt.plot(T_x, S_x)
    plt.plot(T_y, S_y)
    plt.xlabel('t')
    plt.ylabel('s')
    
    plt.figure(2)
    ax = plt.gca()
    ax.set_aspect('equal')
    ax.grid(True, which='both')
    plt.plot(X[:, 0], Y[:, 0])
    delta = 0.1
    plt.plot(x_10, y_10, '.c')
    plt.plot(x_1f, y_1f, '.g')
    plt.plot([x_10, x_10 + x_20*delta], [y_10, y_10 + y_20*delta], '-c')
    plt.plot([x_1f, x_1f + x_2f*delta], [y_1f, y_1f + y_2f*delta], '-g')
    plt.ylim([-2, 2])
    plt.xlim([-2, 2])
    plt.xlabel('x_1')
    plt.ylabel('y_1')
    plt.show()
```

```python
plt.figure(3)
di = DoubleIntegrator2D(u_sys_lim=1.0, dt=0.01)

for i in range(100):
    seed(i)
    x_10 = random()*2.0 - 1.0
    x_20 = random()*2.0 - 1.0
    x_1f = random()*2.0 - 1.0
    x_2f = random()*2.0 - 1.0
    y_10 = random()*2.0 - 1.0
    y_20 = random()*2.0 - 1.0
    y_1f = random()*2.0 - 1.0
    y_2f = random()*2.0 - 1.0
    
    res, traj = di.steer(x_10=x_10, x_20=x_20, x_1f=x_1f, x_2f=x_2f, y_10=y_10, y_20=y_20, y_1f=y_1f, y_2f=y_2f)

    if res:
        [t_f_x, t_s_x, y_1s, y_2s, X, U_x, S_x, T_x, t_f_y, t_s_y, y_1s, y_2s, Y, U_y, S_y, T_y] = traj

        ax = plt.gca()
        ax.set_aspect('equal')
        ax.grid(True, which='both')
        plt.plot(X[:, 0], Y[:, 0])
        delta = 0.1
        plt.plot(x_10, y_10, '.c')
        plt.plot(x_1f, y_1f, '.g')
        plt.plot([x_10, x_10 + x_20 * delta], [y_10, y_10 + y_20 * delta], '-c')
        plt.plot([x_1f, x_1f + x_2f * delta], [y_1f, y_1f + y_2f * delta], '-g')
        plt.ylim([-2, 2])
        plt.xlim([-2, 2])
        plt.xlabel('x_1')
        plt.ylabel('y_1')
plt.show()
```
